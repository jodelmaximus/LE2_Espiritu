```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
car_evaluation = fetch_ucirepo(id=19)

# data (as pandas dataframes)
X = car_evaluation.data.features
y = car_evaluation.data.targets

# metadata
print(car_evaluation.metadata)

# variable information
print(car_evaluation.variables)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.feature_selection import RFE
from sklearn.metrics import accuracy_score
```

```python
models = {
    'Logistic Regression': make_pipeline(StandardScaler(), LogisticRegression(random_state=42, multi_class='auto')),
    'Random Forest': RandomForestClassifier(random_state=42),
    'Support Vector Classifier': make_pipeline(StandardScaler(), SVC(kernel='rbf', probability=True, random_state=42)),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)
}
```

```python
def plot_confusion_matrix(y_true, y_pred, classes, model_name):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, cmap='Blues', fmt='g', xticklabels=classes, yticklabels=classes)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title(f'Confusion Matrix - {model_name}')
    plt.show()

# Define a function to plot classification report
def plot_classification_report(y_true, y_pred, model_name):
    report = classification_report(y_true, y_pred)
    print(f'Classification Report - {model_name}:\n\n{report}')

```

```python
results_all_features = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    accuracy = accuracy_score(y_test, predictions)
    results_all_features[name] = accuracy
    plot_confusion_matrix(y_test, predictions, label_encoder.classes_, name)

# Plot the metrics in a bar chart
plt.figure(figsize=(10, 6))
plt.bar(results_all_features.keys(), results_all_features.values(), color='skyblue')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Accuracy of Different Models All Features')
plt.ylim(0, 1)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
print(results_all_features)

results_rfe_selected_features = {}
for name, model in models.items():
    model.fit(X_train_selected, y_train)
    predictions = model.predict(X_test_selected)
    accuracy = accuracy_score(y_test, predictions)
    results_rfe_selected_features[name] = accuracy
    plot_confusion_matrix(y_test, predictions, label_encoder.classes_, name)

# Plot the metrics in a bar chart
plt.figure(figsize=(10, 6))
plt.bar(results_all_features.keys(), results_all_features.values(), color='skyblue')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Accuracy of Different Models Selected Features')
plt.ylim(0, 1)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
print(results_rfe_selected_features)

results_df = pd.DataFrame({
    "Model": list(results_all_features.keys()) + list(results_rfe_selected_features.keys()),
    "Accuracy": list(results_all_features.values()) + list(results_rfe_selected_features.values()),
    "Feature Selection": ['All Features'] * len(models) + ['RFE Selected Features'] * len(models)
})

plt.figure(figsize=(12, 8))
sns.barplot(data=results_df, x='Model', y='Accuracy', hue='Feature Selection')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Model Accuracy Comparison')
plt.xticks(rotation=45)
plt.ylim(0.5, 1.0)
plt.legend(title='Feature Selection')
plt.tight_layout()
plt.show()
```

```python
best_model_name = max(results_all_features, key=results_all_features.get)
best_model = models[best_model_name]

# Train the best model
best_model.fit(X_train, y_train)

# Predict probabilities for the positive class (class 1) on test data
y_test_probabilities = best_model.predict_proba(X_test)[:, 1]
print("The best model is:", best_model_name)
```

```python
from sklearn.ensemble import BaggingClassifier, StackingClassifier

best_model.fit(X_train_selected, y_train)
predictions_highest = best_model.predict(X_test_selected)
accuracy_highest = accuracy_score(y_test, predictions_highest)


stacking_model = StackingClassifier(estimators=[('best_model', SVC(kernel='rbf', probability=True, random_state=42))], final_estimator=LogisticRegression(max_iter=10000))
stacking_model.fit(X_train_selected, y_train)
stacking_accuracy = stacking_model.score(X_test_selected, y_test)

# Bagging Classifier
bagging_model = BaggingClassifier(estimator=best_model, n_estimators=100)
bagging_model.fit(X_train_selected, y_train)
bagging_accuracy = bagging_model.score(X_test_selected, y_test)

models_accuracy = {
    "Original Best Model": accuracy_highest,
    "Stacking": stacking_accuracy,
    "Bagging": bagging_accuracy
}

plt.figure(figsize=(10, 6))
sns.barplot(x=list(models_accuracy.keys()), y=list(models_accuracy.values()))
plt.xlabel('Ensemble Technique')
plt.ylabel('Accuracy')
plt.title('Model Accuracy Comparison Before and After Ensemble Techniques')
plt.ylim(0.0, 1.0)
plt.tight_layout()
plt.show()
for model, accuracy in models_accuracy.items():
    print(f"{model} Accuracy: {accuracy}")
```
